```python
df = df.dropna()
```

```python
import numpy as np
df[["Strength_Att", "Strength_Def"]] = df[["Strength_Att", "Strength_Def"]].astype(np.float).astype("Int64")
```

```python
df.dtypes
```

```python
df.at[8, "Strength_Def"] = 4594 # Battle of Buena Vista Wikipedia
```

```python
df.head(10)
```

```python
df["Att_to_Def"] = round(df["Strength_Att"]/ df["Strength_Def"], 2)
```

```python
df.head(10)
```

```python
df.insert(6, column = "Victor", value = ["A", "D", "A", "A", "A", "A", "A", "D", "A", "D", "D", "D", "D", "D", "A", "D", "A", 
                                        "A", "A", "A", "A", "A", "D", "A", "D", "D", "A", "A", "A", "A", "A", "A", "A", "A", 
                                         "D", "D", "D", "A", "A", "A", "A", "A"])
```

```python
df.head(10)
```

```python
print("In 42 battles,", df.Victor.value_counts()[0], " attackers were successful.")
print("In 42 battles,", df.Victor.value_counts()[1], " defenders were successful.")
```

```python
print(len(df[df.Strength_Att < df.Strength_Def].Victor), "numerically inferior attackers;",
      df[df.Strength_Att < df.Strength_Def].Victor.value_counts()[0], "were successful.")
```

```python
print(len(df[(df.Victor == "A") & (df.Strength_Att > df.Strength_Def) | 
             (df.Victor == "D") & (df.Strength_Att < df.Strength_Def)]), "victors were numerically superior.")
```

```python
print(len(df[(df.Victor == "A") & (df.Strength_Att < df.Strength_Def) | 
       (df.Victor == "D") & (df.Strength_Att > df.Strength_Def)]), "victors were numerically inferior.")
```
