**Notes regarding the Conflict Catalog**
Peter Brecke, 20 August 2012
The file “Conflict Catalog 18 vars” is an Excel (.xls) file and is also the version of the dataset I have sent to those who have asked for the Conflict Catalog.<br>
The file should be available at:http://www.cgeh.nl/institutions-hub<br>
A description of the Conflict Catalog can be found in a Peace Science Society conference paper available at:<br>
http://www.inta.gatech.edu/peter/PSS99_paper.html<br>
If you use the Conflict Catalog, please use that paper for a citation.<br>
It must be understood that the Conflict Catalog within that Excel file is an unfinished product. <br>
Taking into consideration the data I have in my office that is not yet in the Conflict Catalog, I can easily believe the catalog growing by 20% or more.<br>
There are certainly errors in the data, especially as we go back in time and into particular regions of the world.<br>
I have strived mightily to make these data as correct and accurate as I can, but there are undoubtedly errors of some form or other in the file.<br>
I tried to eliminate errors and resolve inconsistencies by comparing as many sources as I could find for every datapoint in the file.<br>
I believe the number of mistakes in terms of typos is quite small.<br>
Within the Name field, conflicts between political entities possessing effective sovereignty (as opposed to sovereignty recognized by others) over different territories
are demarcated by a hyphen between the sets of protagonists, those who deploy fighting forces.<br>
These are effectively interstate wars. I, whenever possible, designated the initiator by placing it first in the list of participants.<br>
Conflicts primarily within a political entity possessing effective sovereignty, namely primarily between a state and some set of sub-national actors, have the sub-national
actors in parentheses following the state name.<br>
These include overthrow rebellions and separatist rebellions as well as other types of domestic political conflict.<br>
Start and end dates go to the precision of which day when possible.<br>
In some instances my dates differ from most sources. I will stand behind my dates. They were the best I could do using as a consistent rule for the starting and ending dates that those are the
dates when the weapons begin to clash and when they cease.<br>
Fatalities (not casualties) are distinguished in terms of military, civilian, and total.<br>
"Military fatalities" is best thought of as soldiers killed in battles.<br>
"Total fatalities"encompasses everyone.<br>
The military fatalities estimates, I hope, are reasonably close to being just soldiers, but I expect there are some civilians in the counts.<br>
The total fatalities estimates, I am sure, vary significantly in terms of the degree to which they include disease, starvation, etc.<br>
The Conflict Catalog contains a few examples of conflicts consisting of meta-wars and sub-wars. The examples of meta-wars containing sub-wars I can think of are the
Hundred Years War, the War of the Roses, the Dutch rebellion against Spain, the 30 Years War, the Napoleonic wars, WW1, and WW2. <br>
For each, if I could discern fatalities for a sub-war, I would put that information into the Conflict Catalog.<br>
The data can be sorted by region. The regions are specified in the online paper.<br>
The file effectively contains the following information:<br>
- Who fought<br>
- Who was on each side in > 2 participant conflicts (separated by hyphens)<br>
- Who initiated (first actor to left of hyphen) (done wherever possible)<br>
- Year started<br>
- Month started (when possible)<br>
- Day started (when possible)<br>
- Year ended<br>
- Month ended (when possible)<br>
- Day ended (when possible)<br>
- Location (when possible)<br>
- Common name of conflict<br>
- Number of military fatalities (when possible)<br>
- Total number of fatalities (when possible)<br>
- Number of significant political actors (not number of combatants in terms of number of soldiers)<br>
The reason for the 1400 start date is that was Evan Luard's start date, and his conflict compilation was the one that formed the real starting point for my effort.<br>
My experience has been that for areas of the world other than Europe, China, Japan, and to a degree
India, attempting to go earlier than 1400 runs into a big drop-off in availability of data.<br>
The historical record I have found simply picks up only the big wars, if any.<br>
I have a dataset of European conflicts from 900-1399 that extends the Conflict Catalog back for regions 3 and 4 (West and East Europe).<br>
While quite extensive (1127 conflicts), it is also not a finished product.<br>
I have from separate work capabilities data for 14 countries going back as far as to 1494, but they have not been integrated here for calculating a ratio of capabilities.<br>
I also have population data that would enable calculation of weighted fatalities, as well as fatality concentration and fatality density variables, but they too have not been
integrated into the conflict catalog.<br>
There are more conflicts for which I have fatalities data, but they have not been integrated either.<br>
